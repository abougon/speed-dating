```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data1, data2 = np.random.randn(2,1000) #generer 2 arrays de 1000 nombres normalement distribues
```

```python
data3 = pd.DataFrame({"Colonne1":[1,2,3,4], "Colonne2":[9,8,7,6]})
data3.head()
```

```python
fig, ax = plt.subplots()
ax.plot(data3.Colonne1, data3.Colonne2, color="C3", marker="x",linestyle='--')
```

```python
fig, axes = plt.subplots(3,3)
fig.set_size_inches(15,15)

#HISTOGRAMME:DISTPLOT
n = axes[0,0].hist(data1, bins=100, color="C1") #Colors => C0  C9 

style = dict(color="C8", linewidth=2, linestyle="--", marker="o")
style2 = dict(color="C2", marker="x")

#LINEPLOT : #equivalent de relplot avec kind = line (lineplot : nuage des points relis entre 2 par une ligne pour visualiser des relations lineaires)
axes[0,1].plot(data3["Colonne1"], data3["Colonne2"], **style)

#SCATTERPLOT : #equivalent de relplot avec kind = scatter (scatter = nuage de points)
axes[0,2].scatter(data3["Colonne2"], data3["Colonne1"], marker="x", color="C5")

#KDEPLOT : kernel density estimation #noyau dur de vos donnes
sns.kdeplot([1,2,3,4,5,6,7],[22,12,56,73,23,78,53], shade=True, shade_lowest=True, ax=axes[1,0])

#BARPLOT :#equivalent du catplot kind = bar #vertical ou horizontal
axes[1,1].bar(["voitures", "camions", "motos", "velos"], [2,3,1,5], color="C2")

#STACKED BAR : Visualiser une var categorique en fonction de 2 variables num
axes[1,2].bar(["voitures", "camions", "motos", "velos"], [2,3,1,5], color="C2")
axes[1,2].bar(["voitures", "camions", "motos", "velos"], [6,7,8,9], bottom = [2,3,1,5])
axes[1,2].bar(["voitures", "camions", "motos", "velos"], [6,7,8,9], bottom = [8,10,9,14])

axes[2,2].bar(["voitures", "camions", "motos", "velos"], [2,3,1,5], color="C2")
axes[2,2].bar(["voitures", "camions", "motos", "velos"], [6,7,8,9], color="C7")


#PIE CHART : camembert
m = axes[2,0].pie([2,3,1,5], labels=["voitures", "camions", "motos", "velos"], autopct="%1.0f%%", explode=(0, 0.15, 0, 0))
axes[2,0].set_title("Repartition vehicules", fontsize = 15)

#BOXPLOT : bote  moustaches
axes[2,1].boxplot(data1, showmeans=True, meanline=True)
plt.grid()
plt.show()
```

```python
#STACKED BAR :

countries = ['Norway', 'Germany', 'Canada', 'United States', 'Netherlands']
bronzes = np.array([10,7,10,6,6])
silvers = np.array([14,10,8,8,6])
golds = np.array([14,14,11,9,8])
ind = [country for country in countries]
plt.bar(ind, golds, width=0.6, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar(ind, silvers, width=0.6, label='silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.6, label='bronzes', color='#CD7F32')
plt.xticks(ind, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right")
plt.title("2018 Winter Olympics Top Scorers")
plt.show()
```

```python
x = np.random.randn(100)
```

```python
y = np.sort(x)
```

```python
y
```

```python
Q1=y[24] #25e element de mon array
Q3=y[74] #75e element de mon array
```

```python
whis=1.5
```

```python
Upper_bound = Q3 + whis*(Q3-Q1)
Lower_bound = Q1 - whis*(Q3-Q1)
```

```python
Upper_bound, Lower_bound
```

```python
Q1, Q3
```

```python
Q3 + whis*(Q3-Q1), Q1 - whis*(Q3-Q1)
```

```python
whis = 1.5 #whis=1.5 98% de confiance #whis=1 95% intervalle de confiance
```

```python
Q3 + whis*(Q3-Q1), Q1 - whis*(Q3-Q1)
```

```python
#Apparatre tes outliers en jouant sur le whis:

for i in x:
    if i > Upper_bound or i < Lower_bound:
        print(i)
```

```python
plt.boxplot(x, whis=whis, showmeans=True)
```

```python
names = ["voitures", "motos", "scooter"]
values = [1,10,100]
values2 = [7,3,10]

#Creation d'un barplot (stacked bar)

plt.figure(figsize=(10,10)) #par defaut figure de taille (8,6)
plt.bar(names, values)
plt.bar(names, values2, bottom=values)
plt.title("BARCHART_1")
plt.legend(["valeurs1", "valeurs2"]) #liste de vos variables numeriques => renommer 
plt.xlabel("Nom Abscisses")
plt.ylabel("Nom Ordonnees")
plt.text(names[0],2, values[0]) #hauteur de mon texte sur la barre
plt.text(names[0],10, values2[0])
plt.text(names[1],20, values[1]+values2[1])
plt.text(names[2],100, values[2]+values2[2])

plt.show()
```

```python
#Plusieurs graphes dans une seule figure

#Creation de donnes alatoires sinusoidales

t = np.arange(0,2,0.01)
s = np.sin(2*np.pi*t) #equation de la fonction sin
u = 1 + np.log(2*t) #equation de la fonction log

#Creation de lineplots

plt.plot(t,s, linestyle='--')
plt.plot(t,u)
plt.grid(True)
plt.show()
```
